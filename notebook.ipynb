Question 1 - Extracting Tesla Stock Data Using yfinance - 2 Points

```python
import yfinance as yf

# Download historical data for Tesla
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="max")

# Display the first few rows of the data
print(tesla_data.head())
```

Question 2 - Extracting Tesla Revenue Data Using Webscraping - 1 Points

```python
import requests
from bs4 import BeautifulSoup

# URL da página com os dados de receita da Tesla
url = "URL_DA_PAGINA_COM_DADOS_DE_RECEITA"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Exemplo: Encontrar a receita em uma tabela
# (A estrutura real pode variar dependendo da página web)
table = soup.find('table', {'class': 'classe_da_tabela'})
rows = table.find_all('tr')
for row in rows:
    cols = row.find_all('td')
    for col in cols:
        print(col.text)
```
